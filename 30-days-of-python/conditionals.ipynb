```python
a = 22
if a > 0:
    if a % 2 == 0:
        print("A is a positive integer and an even number.")
    else:
        print("A is a positive integer.")
elif a < 0:
    print("A is a negative integer.")
else:
    print("A is zero.")
```

```python
a = 0
if a > 0 and a % 2 == 0:
    print("A is a positive integer and an even number.")
elif a > 0 and a % 2 != 0:
    print("A is a positive integer and an odd number.")
elif a == 0:
    print("A is zero.")
else:
    print("A is a negative integer.")
```

```python
age = int(input("Enter your age: "))
if age >= 18:
    print("You are old enough to learn to drive.")
else:
    print(f"You need {18 - age} more years to learn to drive.")
```

```python
my_age = 22
your_age = int(input("Enter your age: "))

if my_age > your_age:
    if my_age - your_age == 1:
        print(f"I'm {my_age - your_age} year older than you.")
    else:
        print(f"I'm {my_age - your_age} years older than you.")
elif my_age < your_age:
    if your_age - my_age == 1:
        print(f"You are {your_age - my_age} year older than me.")
    else:
        print(f"You are {your_age - my_age} years older than me.")
else:
    print("We are of the same age.")
```
