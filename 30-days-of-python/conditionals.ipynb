```python
a = 22
if a > 0:
    if a % 2 == 0:
        print("A is a positive integer and an even number.")
    else:
        print("A is a positive integer.")
elif a < 0:
    print("A is a negative integer.")
else:
    print("A is zero.")
```

```python
a = 0
if a > 0 and a % 2 == 0:
    print("A is a positive integer and an even number.")
elif a > 0 and a % 2 != 0:
    print("A is a positive integer and an odd number.")
elif a == 0:
    print("A is zero.")
else:
    print("A is a negative integer.")
```
