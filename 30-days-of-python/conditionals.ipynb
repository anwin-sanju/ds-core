```python
age = int(input("Enter your age: "))
if age >= 18:
    print("You are old enough to learn to drive.")
else:
    print(f"You need {18 - age} more years to learn to drive.")
```

```python
my_age = 22
your_age = int(input("Enter your age: "))

if my_age > your_age:
    if my_age - your_age == 1:
        print(f"I'm {my_age - your_age} year older than you.")
    else:
        print(f"I'm {my_age - your_age} years older than you.")
elif my_age < your_age:
    if your_age - my_age == 1:
        print(f"You are {your_age - my_age} year older than me.")
    else:
        print(f"You are {your_age - my_age} years older than me.")
else:
    print("We are of the same age.")
```

```python
a = int(input("Enter a: "))
b = int(input("Enter b: "))

if a > b:
    print("A greater than B")
elif a < b:
    print("B greater than A")
else:
    print("A and B are same")
```

```python
score = int(input("Enter your score: "))

if score >= 90 and score <= 100:
    print("A Grade")
elif score >= 80:
    print("B Grade")
elif score >= 70:
    print("C Grade")
elif score >= 60:
    print("D Grade")
elif score < 60 and score >= 0:
    print("F Grade")
else:
    print("Invalid Score")
```

```python
month = input("Enter the month").title()

if month in ['September', 'October', 'November']:
    print("Season is Autumn")
elif month in ['December', 'January', 'February']:
    print("Season is Winter")
elif month in ['March', 'April', 'May']:
    print("Season is Spring")
elif month in ['June', 'July', 'August']:
    print("Season is Summer")
else:
    print("Invalid month")
```

```python
fruits = ['banana', 'orange', 'mango', 'lemon']

fruit = input("Enter a fruit name: ").lower()

if fruit in fruits:
    print(fruits)
    print(f"{fruit} already exists.")
else:
    fruits.append(fruit)
    print(fruits)
```
