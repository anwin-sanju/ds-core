## Notes

```python
# syntax
# [expression for i in iterable if condition]
```

```python
language = "Python"

lst = [i for i in language]
print(lst)
```

```python
numbers = [i for i in range(11)]
print(numbers)
```

```python
squares = [i * i for i in range(11)]
print(squares)
```

```python
list_of_squares = [(i, i * i) for i in range(11)]
print(list_of_squares)
```

```python
even_numbers = [i for i in range(21) if i % 2 == 0]
print(even_numbers)
```

```python
odd_numbers = [i for i in range(21) if i % 2 != 0]
print(odd_numbers)
```

```python
numbers = [-8, -7, -3, -1, 0, 1, 3, 4, 5, 7, 6, 8, 10]
positive_even_numbers = [i for i in numbers if i % 2 == 0 and i > 0]
print(positive_even_numbers)
```

```python
list_of_lists = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened_list = [number for row in list_of_lists for number in row]
print(flattened_list)
```

### syntax (lambda functions)
x = lambda param1, param2, param3: param1 + param2 + param3
print(x(arg1, arg2, arg3))

```python
add_two_nums = lambda a, b: a + b
print(add_two_nums(2, 4))
```

```python
print((lambda a, b: a + b)(2, 4))
```

```python
square = lambda x: x ** 2
print(square(4))
```

```python
cube  = lambda x: x ** 3
print(cube(3))
```

```python
multiple_variable = lambda a, b, c: a ** 2 - 3 * b + 4 * c
print(multiple_variable(5, 5, 3))
```

```python
def power(x):
    return lambda n: x ** n

power(5)(3)
```

```python
cube = power(2)(3)
print(cube) 
```

```python
two_power_of_five = power(2)(5) 
print(two_power_of_five)
```

```python
numbers = [-4, -3, -2, -1, 0, 2, 4, 6]
negative_and_zero = [i for i in numbers if i <= 0]
print(negative_and_zero)
```

```python
list_of_lists =[[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [number for row in list_of_lists for number in row]
print(flat)
```

```python
list_of_square = [(i, i**0, i**1, i**2, i**3, i**4, i**5) for i in range(11)]
print(list_of_square)
```

```python
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]
```

```python
print(countries[0])
```

```python
countries[0][0]
```

```python
flattened = [[item[0].upper(), item[0].upper()[:3], item[1].upper()] for row in countries for item in row]
print(flattened)
```

```python
flattened_dict = [{'country': item[0].upper(), 'city': item[1].upper()} for row in countries for item in row]
print(flattened_dict)
```

```python
names = [[('Asabeneh', 'Yetayeh')], [('David', 'Smith')], [('Donald', 'Trump')], [('Bill', 'Gates')]]
```

```python
name_list = [f"{item[0]} {item[1]}" for name in names for item in name]
print(name_list)
```

```python
y_intercept = lambda x, y, m: y - (m * x)
slope = lambda x1, y1, x2, y2: (y2 - y1) / (x2 - x1)

print(y_intercept(2, 7, 2))
print(slope(1, 2, 3, 6))
```
