```python
count = 5
while count != 0:
    print(count)
    count -= 1
else:
    print(count)
```

```python
count = 0
while count < 5:
    print(count)
    count += 1
    if count == 3:
        break
```

```python
count = 0
while count < 5:
    if count == 3:
        count += 1
        continue
    print(count)
    count += 1
```
