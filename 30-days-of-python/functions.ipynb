```python
import math
def area_of_circle(radius):
    return math.pi * radius ** 2

area_of_circle(30)
```

```python
def sum_of_numbers(n):
    total = 0
    for i in range(n + 1):
        total += i
    return total

sum_of_numbers(50)
```

```python
def calculate_age(current_year, birth_year):
    return current_year - birth_year

calculate_age(2026, 2004)
```

```python
def weight_of_object(mass, gravity):
    return str(mass * gravity) + ' N'

weight_of_object(100, 9.81)
```

```python
def add(num1, num2):
    return num1 + num2

add(num2=4, num1=5)
```

```python
def is_even(n):
    if int(n) % 2 == 0:
        return True
    return False

is_even("358787689347564")
```

```python
def find_even_numbers(n):
    evens = []
    for num in range(n + 1):
        if num % 2 == 0:
            evens.append(num)
    return evens

find_even_numbers(20)
```

```python
def sum_all_nums(*nums):
    total = 0
    for num in nums:
        total += num
    return total

sum_all_nums(2, 4, 6)
```

```python
def team_details(name, *players):
    print(name)
    for player in players:
        print(player)

team_details("Hogwarts", "Hazel", "Bellatrix")
```
