```python
import math

def solve_quadratic_eqn(a, b, c):
    discriminant = b**2 - 4*a*c

    if discriminant > 0:
        x1 = (-b + math.sqrt(discriminant)) / (2 * a)
        x2 = (-b - math.sqrt(discriminant)) / (2 * a)
        return x1, x2

    elif discriminant == 0:
        x = -b / (2 * a)
        return x

    else:
        return "No real solutions"

print(solve_quadratic_eqn(1, -3, 2))
```

```python
print(solve_quadratic_eqn(1, 2, 1))
```

```python
def sum_of_odds(n):
    odd_sum = 0
    for i in range(n + 1):
        if i % 2 != 0:
            odd_sum += i
    return odd_sum

sum_of_odds(10)
```

```python
def sum_of_evens(n):
    even_sum = 0
    for i in range(0, n + 1, 2):
        even_sum += i
    return even_sum

sum_of_evens(10)
```

```python
def evens_and_odds(n):
    even_count = 0
    odd_count = 0
    for i in range(n + 1):
        if i % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
    return even_count, odd_count

even_count, odd_count = evens_and_odds(100)
print(f"The number of odds are {odd_count}")
print(f"The number of evens are {even_count}")
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

factorial(5)
```

```python
def calculate_mean(lst):
    return sum(lst) / len(lst)

def calculate_median(lst):
    sorted_list = sorted(lst)
    mid = len(lst) // 2

    if len(lst) % 2 == 0:
        median = (sorted_list[mid - 1] + sorted_list[mid]) / 2
    else:
        median = sorted_list[mid]
    
    return median
```

```python
numbers = [1, 2, 2, 3, 4, 4, 4, 5]
calculate_mean(numbers)
```

```python
calculate_median(numbers)
```

```python
def mode(lst):
    item_count = {}
    for item in lst:
        if item in item_count:
            item_count[item] += 1
        else:
            item_count[item] = 1

    max_count = 0
    mode_value = None
    for k, v in item_count.items():
        if v > max_count:
            max_count = v
            mode_value = k
    return mode_value
            

numbers = [10, 20, 20, 30, 40, 50, 50, 50, 60]

print(mode(numbers))
```

```python
def calc_range(lst):
    return max(lst) - min(lst)

numbers = [10, 20, 20, 30, 40, 50, 50, 50, 60]
calc_range(numbers)
```

```python
def calc_variance(lst):
    mean = calculate_mean(lst)
    variance_sum = 0
    for number in lst:
        variance_sum += (number - mean) ** 2
    return variance_sum / len(lst)

numbers = [10, 20, 20, 30, 40, 50, 50, 50, 60]
calc_variance(numbers)
```

```python
import math
def calc_std(lst):
    variance = calc_variance(lst)
    return math.sqrt(variance)

numbers = [10, 20, 20, 30, 40, 50, 50, 50, 60]
calc_std(numbers)
```
