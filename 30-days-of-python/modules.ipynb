```python
import random
import string

def random_user_id():
    characters = string.digits + string.ascii_lowercase
    word = ""
    for i in range(6):
        word += random.choice(characters)
    return word

random_user_id()
```

```python
def user_id_gen_by_user():
    no_of_characters = int(input("Enter the no. of characters: "))
    no_of_id = int(input("Enter the no. of ID's: "))

    characters = string.digits + string.ascii_letters
    
    for i in range(no_of_id):
        word = ""
        for j in range(no_of_characters):
            word += random.choice(characters)
        print(word)

user_id_gen_by_user()
```

```python
def rgb_color_gen():
    red = random.randint(0, 255)
    green = random.randint(0, 255)
    blue = random.randint(0, 255)
    rgb = (red, green, blue)
    return f"rgb{rgb}"

print(rgb_color_gen())
```

```python
def list_of_hexa_colors():
    
    n = int(input("No. of Hexadecimal numbers: "))
    colors = []
    characters = "0123456789abcdef"

    for i in range(n):
        user_id = "#"
        for j in range(6):
            user_id += random.choice(characters)
        colors.append(user_id)
        print(user_id)
    return colors

list_of_hexa_colors()
```

```python
import random

def list_of_rgb_colors():
    
    n = int(input("No. of colors: "))
    colors = []

    for i in range(n):
        r = random.randint(0, 255)
        g = random.randint(0, 255)
        b = random.randint(0, 255)
        
        color = f"rgb({r},{g},{b})"
        colors.append(color)

    return colors

print(list_of_rgb_colors())
```

```python
def generate_colors(color_type):
    if color_type == "hexa":
        list_of_hexa_colors()
    elif color_type == "rgb":
        list_of_rgb_colors()
    else:
        return False
```

```python
generate_colors('hexa') 
```

```python
def shuffle_list(items):
    new_list = items.copy()
    random.shuffle(new_list)
    return new_list

numbers = [1, 2, 3, 4, 5]
print(shuffle_list(numbers))
```

```python
def unique_random_numbers():

    numbers = []
    
    while len(numbers) < 7:
        number = random.randint(0, 9)
        if number not in numbers:
            numbers.append(number)
    
    return numbers

unique_random_numbers()
```
